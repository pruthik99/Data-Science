```python
import numpy as np
import matplotlib.pyplot as mlb
```

# Normal Distribution

```python
fig, a = mlb.subplots(4,4,figsize = (16,15))
mu = 1
sd = 1
sd1 = 4
sd2 = 10
sd3 = 100
x = np.random.normal(mu,sd,1)
x1 = np.random.normal(mu,sd1,10)
x2 = np.random.normal(mu,sd2,100)
x3 = np.random.normal(mu,sd3,1000)
a[0][0].hist(x,bins = 40)
a[1][0].hist(x1,bins = 40)
a[2][0].hist(x2,bins = 40)
a[3][0].hist(x3,bins = 40)
a[0][0].set_title('n = 1 and shape = 1')
a[1][0].set_title('n = 10 and shape = 1')
a[2][0].set_title('n = 100 and shape = 1')
a[3][0].set_title('n = 1000 and shape = 1')
a[0][1].hist(x,bins = 40)
a[1][1].hist(x1,bins = 40)
a[2][1].hist(x2,bins = 40)
a[3][1].hist(x3,bins = 40)
a[0][1].set_title('n = 1 and shape = 4')
a[1][1].set_title('n = 10 and shape = 4')
a[2][1].set_title('n = 100 and shape = 4')
a[3][1].set_title('n = 1000 and shape = 4')
a[0][2].hist(x,bins = 40)
a[1][2].hist(x1,bins = 40)
a[2][2].hist(x2,bins = 40)
a[3][2].hist(x3,bins = 40)
a[0][2].set_title('n = 1 and shape = 10')
a[1][2].set_title('n = 10 and shape = 10')
a[2][2].set_title('n = 100 and shape = 10')
a[3][2].set_title('n = 1000 and shape = 10')
a[0][3].hist(x,bins = 40)
a[1][3].hist(x1,bins = 40)
a[2][3].hist(x2,bins = 40)
a[3][3].hist(x3,bins = 40)
a[0][3].set_title('n = 1 and shape = 100')
a[1][3].set_title('n = 10 and shape = 100')
a[2][3].set_title('n = 100 and shape = 100')
a[3][3].set_title('n = 1000 and shape = 100')
```

# Exponential Distribution

# Binomial Distribution

```python
fig, c = mlb.subplots(4,4,figsize = (16,15))
x = np.random.binomial(n = 1,p = 0.2, size = 1)
x1 = np.random.binomial(n = 4, p = 0.3, size = 10)
x2 = np.random.binomial(n = 10, p = 0.4, size = 100)
x3 = np.random.binomial(n = 100, p = 0.5, size = 1000)
c[0][0].hist(x,bins = 40)
c[1][0].hist(x1,bins = 40)
c[2][0].hist(x2,bins = 40)
c[3][0].hist(x3,bins = 40)
c[0][0].set_title('n = 1 and shape = 1')
c[1][0].set_title('n = 10 and shape = 1')
c[2][0].set_title('n = 100 and shape = 1')
c[3][0].set_title('n = 1000 and shape = 1')
c[0][1].hist(x,bins = 40)
c[1][1].hist(x1,bins = 40)
c[2][1].hist(x2,bins = 40)
c[3][1].hist(x3,bins = 40)
c[0][1].set_title('n = 1 and shape = 4')
c[1][1].set_title('n = 10 and shape = 4')
c[2][1].set_title('n = 100 and shape = 4')
c[3][1].set_title('n = 1000 and shape = 4')
c[0][2].hist(x,bins = 40)
c[1][2].hist(x1,bins = 40)
c[2][2].hist(x2,bins = 40)
c[3][2].hist(x3,bins = 40)
c[0][2].set_title('n = 1 and shape = 10')
c[1][2].set_title('n = 10 and shape = 10')
c[2][2].set_title('n = 100 and shape = 10')
c[3][2].set_title('n = 1000 and shape = 10')
c[0][3].hist(x,bins = 40)
c[1][3].hist(x1,bins = 40)
c[2][3].hist(x2,bins = 40)
c[3][3].hist(x3,bins = 40)
c[0][3].set_title('n = 1 and shape = 100')
c[1][3].set_title('n = 10 and shape = 100')
c[2][3].set_title('n = 100 and shape = 100')
c[3][3].set_title('n = 1000 and shape = 100')
```
