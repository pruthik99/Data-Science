```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plot
import math
import numpy as np
```

```python
#defining function for calculating minimum and maximum ages
def getMinMaxAge(minAge, maxAge):
    
    #normalizing minimum age to multiplication of 10 plus 1
    minAge = minAge if (minAge % 10) == 1 else ((10 * math.trunc(minAge / 10)) + 1)
    
    #normalizing maximum age to multiplication of 5
    maxAge = maxAge if (maxAge % 5) == 0 else (5 * (math.trunc(maxAge / 5) + 1))
    
    #return min and max age as list
    return [minAge, maxAge]
```

```python
#defining function for dividing age in groups
def getAgeGroups(min_range, max_range):
    
    #defining list for storing age groups
    age_group = []
    
    #iterating age from min to max for defining group in 5 years difference
    for age in range(min_range, max_range + 1, 5):
        age_group.append(f'{age}-{age+4}')
    
    #returning list of age groups
    return age_group
```

```python
#defining function for classification of salary in avg, min and max
def getSalaryClassification(df, age_groups, cl_type):
    
    #defining list for storing salary classification
    cl_salary = []
    
    #iterating all age groups
    for grp in age_groups:
        
        #split age group for getting min and max age of group
        ages = grp.split('-')
        
        #selecting salary column
        clmn = df.loc[(df['age'] >= int(ages[0])) & (df['age'] <= int(ages[1])), 'salary']
        
        #calculate average salary
        if cl_type == 'avg':    
            cl_salary.append(round(clmn.mean(), 2))
        
        #calculate maximum salary
        elif cl_type == 'max':
            cl_salary.append(round(clmn.max(), 2))
            
        #calculate minimum salary
        elif cl_type == 'min':
            cl_salary.append(round(clmn.min(), 2))
        
    #returning list of calculated salaries
    return cl_salary
```

```python
#defining function for generating dataframe from list of values
def getDataframe(age_group, avg_value, min_value, max_value):
    
    #creating dataframe with columns
    df = pd.DataFrame(columns=['group','avg','min','max'])
    
    #assigning age group to df's group column
    df['group'] = age_group
    
    #assigning average salary to df's avg column
    df['avg'] = avg_value
    
    #assigning minimum salary to df's min column
    df['min'] = min_value
    
    #assigning maximum salary to df's max column
    df['max'] = max_value
    
    #returning dataframe
    return df
```

```python
#defining function for displaying bar graph for visualization
def visualizeData(age_group, df, title):
    
    #generating range of values from 0 to length of degrees array
    xAx = np.arange(len(age_group))

    #setting figure size
    graph = plot.figure(figsize = (15, 8))

    #assigning data of SIN and COSIN values to bar chart  
    plot.bar(xAx - 0.12, df['min'], width = 0.25, label = 'Minimum')
    plot.bar(xAx + 0.12, df['avg'], width = 0.25, label = 'Average')
    plot.bar(xAx + 0.36, df['max'], width = 0.25, label = 'Maximum')

    #setting degrees location on x axes
    plot.xticks(xAx, age_group)

    #assigning label as degrees to x axes
    plot.xlabel("Age Groups")

    #assigning label as values to y axes
    plot.ylabel("Salary")
    
    #setting legend
    plot.legend()

    #setting title
    plot.title(title)
    #displaying bar graph
    plot.show()
```

```python
#checking if main function is executing
if __name__ == ('__main__'):
    
    #reading dataset and generating dataframe using pandas
    df = pd.read_csv('data2.csv')
    
    #converting categorical value to numerical values and generating new column
    df['part'] = df['part.of.germany'].map({'west':0, 'east':1})
    
    #removing categorical column from dataframe
    df.drop('part.of.germany', axis = 1, inplace = True)
    
    #separating dataframe for EAST part from main dataframe
    edf = df.loc[df['part'] == 1, ['age','salary']] 
    
    #separating dataframe for WEST part from main dataframe
    wdf = df.loc[df['part'] == 0, ['age','salary']] 

    #calculating minimum and maximum for EAST part
    e_age_min, e_age_max = getMinMaxAge(edf['age'].min(), edf['age'].max())
    
    #generating age groups for EAST part
    e_ages = getAgeGroups(e_age_min, e_age_max)

    #calculating minimum and maximum for WEST part
    w_age_min, w_age_max = getMinMaxAge(wdf['age'].min(), wdf['age'].max())
    
    #generating age groups for WEST part
    w_ages = getAgeGroups(w_age_min, w_age_max)
    
    #generating dataframe for EAST part dataset
    e_df = getDataframe(e_ages, 
                    getSalaryClassification(edf.copy(), e_ages, 'avg'), 
                    getSalaryClassification(edf.copy(), e_ages, 'min'),
                    getSalaryClassification(edf.copy(), e_ages, 'max'))
    
    #generating dataframe for WEST part dataset
    w_df = getDataframe(w_ages, 
                    getSalaryClassification(wdf.copy(), w_ages, 'avg'), 
                    getSalaryClassification(wdf.copy(), w_ages, 'min'),
                    getSalaryClassification(wdf.copy(), w_ages, 'max'))
    
    #displaying EAST part bar graph
    visualizeData(e_ages, e_df, "East Part of Germany")

    #displaying WEST part bar graph
    visualizeData(w_ages, w_df, "West Part of Germany")
```
